# Ответ на дополнительный вопрос
BatchNorm — это метод нормализации входных данных в моделях с дообучением, в том числе нейронных сетях. Batch normalization позволяет избежать проблемы ковариатного сдвига. Используется как стандартный метод нормализации, поскольку показывает высокую эффективность. Однако в некоторых алгоритмах глубокого обучения, если переобучение долгое и плохо предсказуемое и если есть возможность нормализовать данные по слоям (например, в RNN), используют layer normalization — нормализацию входных данных по слоям.

```python
import numpy as np
import pandas as pd
```

```python
# using kaggle
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

df_train = pd.read_csv('/kaggle/input/train.csv')
df_test = pd.read_csv('/kaggle/input/test.csv')
df_train.shape
```

```python
df_test.head(2)
```

```python
df_train = df_train.sample(10000, random_state=0)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer

tf_transformer = TfidfTransformer(use_idf=False)
count_vect = CountVectorizer()

X_train_counts = count_vect.fit_transform(df_train['title'])
X_train_counts = tf_transformer.fit_transform(X_train_counts)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(random_state=123)
classifier.fit(X_train_counts.toarray(), df_train['target']) 
```

```python
from sklearn.metrics import f1_score
f1_score(df_test['target'], classifier.predict(count_vect.transform(df_test['title'])), average='macro')
```
